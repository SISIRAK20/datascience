```python
import numpy as np
x=np.array([[12,16],[5,10]])
y=np.array([[8,23],[5,8]])
print("...Array x...\n",x,"\n...Array y...\n",y)
print("\n...Greater...\n",np.greater(x,y))
print("\n...Lesser...\n",np.less(x,y))
print("\n...Greater than or equal to...\n",np.greater_equal(x,y))
print("\n...Lesser than or equal to...\n",np.less_equal(x,y))
```
